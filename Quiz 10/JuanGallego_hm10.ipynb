```python
data = loadtxt("munich.txt")
```

```python
datafull = where( (data[:,1]>-50) & (data[:,1]<50))[0]
data1    = where( (data[:,0]>1995) & (data[:,0]<1996))[0]
data2    = where( (data[:,0]>2012) & (data[:,0]<2013))[0]
```

```python
time = data[datafull,0]
T    = data[datafull,1]
T1    = data[data1,1]
T2    = data[data2,1]
```

```python
ti = linspace(1, 365, 365)
```

```python
plot(ti,T1, c="blue", label = "1995")
plot(ti,T2, c="red", label="2012")
legend(loc=0)
xlabel("dia")
ylabel("Temperatura")
```

```python
info = loadtxt("gauss_line.txt")
```

```python
def gaussianalin(x, ampl, x_0, sigma,a,b):
    return a*x + b + ampl*exp( -(x-x_0)**2/sigma )
```

```python
x = info[:,0]
y = info[:,1]
scatter(x,y, label="obs")
legend(loc=0)
```

```python
from scipy.optimize import curve_fit
```

```python
p_opt, p_cov = curve_fit( gaussianalin, x, y)
```

```python
p_opt
```

```python
y_opt = gaussianalin(x, p_opt[0], p_opt[1], p_opt[2], p_opt[3] ,p_opt[4])

scatter(x,y, label="obs", c="yellow")
scatter(x,y_opt, c="green", label="ajuste")
legend(loc=0)
```

```python
data = loadtxt("munich_temperatures_average_with_bad_data.txt")
```

```python
data1 = where( (data[:,1]>-50) & (data[:,1]<50))[0]
```

```python
def temp(t, a, b, c ):
    return a*cos(2*pi*t+b)+c
```

```python
t = data[data1,0]
tem    = data[data1,1]
scatter(t,tem, label="obs")
```

```python
p_opti, p_covi = curve_fit( temp, t, tem)
```

```python
p_opti
```

```python
y_opti = temp(t, p_opti[0], p_opti[1], p_opti[2])

scatter(t,tem, label="obs", c="yellow")
scatter(t,y_opti, c="green", label="ajuste")
legend(loc=0)
```
